#  Part one : Edge detection of conventional image

The purpose of this part is to show you how to extract salient details (edges) from images
which is one of the main processing in computer vision for Autonomous Vehicles.

```python
%matplotlib inline

# Librairies Matplotlib, NumPy et OpenCV
import matplotlib.pyplot as plt
import numpy as np
import cv2   #OpenCV

from sklearn.cluster import MiniBatchKMeans
```

1. Download cameraman.tif image in grayscale format

```python
# Located in Path("datasets/Images/cameraman.tif")
img = cv2.imread("Images/cameraman.tif", cv2.IMREAD_GRAYSCALE)
plt.imshow(img, cmap='gray')
plt.show()
```


2-Apply the canny, sobel edge detector or any other detector of your choice

```python
edges = cv2.Canny(img,100,200)

plt.subplot(121),plt.imshow(img,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
plt.show()
```
